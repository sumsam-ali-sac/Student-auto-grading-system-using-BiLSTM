```python
import os
import tkinter as tk
from tkinter import filedialog,font
from pyftpdlib.authorizers import DummyAuthorizer
from pyftpdlib.handlers import FTPHandler
from pyftpdlib.servers import ThreadedFTPServer


class MyFTPHandler(FTPHandler):
    def on_connect(self):
        self.cwd = os.path.expanduser("C://Users//Sumsam Ali//Untitled Folder 11 - Copy//Essays")



def start_ftp_server():
    try:
        authorizer = DummyAuthorizer()

        # Define a function to read users from a file
        def read_users_from_file(file_name):
            users = []
            with open(file_name) as f:
                for line in f:
                    user_info = line.strip().split(",")
                    users.append({"username": user_info[0], "password": user_info[1], "permissions": user_info[2]})
            return users

        # Read users from the students file and create them with limited permissions
        students = read_users_from_file("C://Users//Sumsam Ali//Untitled Folder 11//students.txt")
        for student in students:
            authorizer.add_user(student["username"], student["password"], os.getcwd(), perm="elaw")

        # Read users from the teachers file and create them with full permissions
        teachers =  read_users_from_file("C://Users//Sumsam Ali//Untitled Folder 11//teachers.txt")
        for teacher in teachers:
            authorizer.add_user(teacher["username"], teacher["password"], os.getcwd(), perm="elradfmwM")
        
        authorizer.add_user("bahadur", "123", os.getcwd(), perm="elradfmwM")
        authorizer.add_user("sumsam", "123", os.getcwd(), perm="elradfmwM")

        handler = MyFTPHandler
        handler.authorizer = authorizer
        server = ThreadedFTPServer(("127.0.0.1", 21), handler)
        server.max_cons = 4 # Maximum number of connections
        server.max_cons_per_ip = 2 # Maximum number of connections per IP address
        server.serve_forever()
    except Exception as e:
        print(f"Error starting FTP server: {e}")

def select_directory():
    try:
        directory = filedialog.askdirectory()
        os.chdir(directory)
        file = open("C://Users//Sumsam Ali//Untitled Folder 11 - Copy//server_directory.txt",'w')
        file.write(directory)
        file.close()
        directory_path_label.config(text=os.getcwd())
    except Exception as e:
        print(f"Error selecting directory: {e}")


if __name__ == '__main__':
    try:
        root = tk.Tk()
        root.geometry("530x200")
        root.configure(bg="#1b1917")
        root.title("FTP Server")
        padding = 10
        # Colors 
        bg_color = "#1b1917" 
        label_color = "#e4e4e4" 
        button_color = "#6e499a" 
        button_text_color = "#fff" 

        # Fonts 
        label_font = font.Font(family="Helvetica", size=12, weight="bold") 
        entry_font = font.Font(family="Helvetica", size=12) 
        button_font = font.Font(family="Helvetica", size=12, weight="bold") 

        # Create a label for the "Directory: " text
        directory_text_label = tk.Label(root, text="Server Directory: ", bg=bg_color, fg=button_color, font=label_font)
        directory_text_label.place(x=50, y=10)

        # Create a label for the directory path
        directory_path_label = tk.Label(root, text=os.getcwd(), bg=bg_color, fg=label_color, font=label_font)
        directory_path_label.place(x=190, y=10)

        select_directory_button = tk.Button(root, text="Select Directory",bg=button_color, fg=button_text_color, font=button_font, command=select_directory ,  width = 15) 
        select_directory_button.place(x=180, y=50)

        start_server_button = tk.Button(root, text="Start Server", bg=button_color, fg=button_text_color, font=button_font,command=start_ftp_server , width = 15) 
        start_server_button.place(x=180, y=100)

        root.mainloop() 
    except Exception as e: 
        print(f"Error starting FTP server GUI: {e}")
```
